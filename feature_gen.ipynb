```python
def conv_features(row):
    features = np.zeros(21, dtype=int)
    pkm = row.squeeze()
    features[type_to_int(pkm["type1"])] = 1
    if (type_to_int(pkm["type2"]) >=0):
        features[type_to_int(pkm["type2"])] = 1
    features[18] = pkm["attack"]+pkm["defense"]+pkm["sp_attack"]+pkm["sp_defense"]+pkm["speed"]+pkm["hp"]
    features[19] = max(pkm["attack"],pkm["sp_attack"]) + pkm["speed"]
    features[20] = max(pkm["defense"],pkm["sp_defense"]) + pkm["hp"]
    return pd.Series(features, index=[f"is_{type_to_int_dict_inv[i]}"for i in range(18)]+['BST', 'ATK_TOT', 'DEF_TOT'])
```

```python
pokemon_features_df = poke.set_index('name').apply(conv_features, axis=1)
pokemon_features_df.to_csv('pokemon_features.csv')
pokemon_features_df
```
